Get rolling monthly betas

```python
def to_returns(prices):
    return prices / prices.shift(1) - 1

def returns_to_index(returns):
    cum_returns = (returns + 1).cumprod()
    return cum_returns.dropna(how='any')

def get_monthly_returns(daily_returns):
    cum_daily_returns = returns_to_index(daily_returns)
    monthly_sample = cum_daily_returns.asfreq('EOM', method='pad')
    return to_returns(monthly_sample).dropna(how='any')

rets = get_monthly_returns(to_returns(close_px))
market = rets.pop('SPX')

def _rolling_beta(returns, benchmark, window):
    model = pd.ols(y=returns,x={'bmk': benchmark}, window=window, min_periods=int(window/2))
    return model.beta['bmk']

def rolling_betas(returns, benchmark, window=250):
    betas = {}
    for col, ts in returns.iteritems():
        betas[col] = _rolling_beta(ts, benchmark, window)
    return pd.DataFrame(betas)

betas = rolling_betas(rets, market, 12)
betas
```
